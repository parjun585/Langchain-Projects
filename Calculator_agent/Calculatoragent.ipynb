```python
from langchain.agents import load_tools
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
llm = ChatOpenAI(temperature=0)
llm2= ChatGroq(
    model_name="groq/compound-mini" 
)

# Load the 'llm-math' tool, which wraps the LLMMathChain
tools = load_tools(["llm-math"], llm=llm2)

```

```python
from langchain import hub
from langchain.agents import create_react_agent, AgentExecutor

# Fetch the ReAct prompt template from LangChain Hub
prompt = hub.pull("hwchase17/react")

# Create the ReAct agent
agent = create_react_agent(llm2, tools, prompt)
```

```python
# Create the Agent Executor
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Define the complex math problem
problem = "If a bakery sells 12 dozen cookies and each cookie costs $0.75, what is the total revenue?"

print(f"--- Running Agent for: {problem} ---")
agent_executor.invoke({"input": problem})
```
